```python
import matplotlib.pyplot as plt
import numpy as np

from sunpy import timeseries as ts
from sunpy.net import Fido
from sunpy.net import attrs as a
```

```python
tstart = "2010-11-05 12:00"
tend = "2010-11-05 17:00"
rg15 = Fido.search(a.Time(tstart, tend), a.Instrument("XRS"), a.goes.SatelliteNumber(15))
print(rg15)
```

```python
# By default downloads to your sunpy/data directory
fg15 = Fido.fetch(rg15)
```

```python
g15 = ts.TimeSeries(fg15)
g15.peek()
```

```python
g15.meta
```

```python
g15.plot()
plt.show()
```
