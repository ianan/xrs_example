```python
#  How different are the Responses for GOES 13 vs GOES15?
```

```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from astropy.io import fits

import goesxrs_temp as gtem
```

```python
resprat15, resptmk15 = gtem.get_resprat(sat=15)
resprat13, resptmk13 = gtem.get_resprat(sat=13)
```

```python
plt.rcParams.update({'font.size': 18,'font.family':"sans-serif",\
                         'font.sans-serif':"Arial",'mathtext.default':"regular"})

# Plot it to see what it is like:
fig,ax = plt.subplots(figsize=(7, 6))
plt.plot(1e6*resptmk15,resprat15,marker=None,color='mediumpurple',label="GOES15",lw=2)
plt.plot(1e6*resptmk13,resprat13,marker=None,color='teal',label="GOES13",lw=2)
ax.set_ylabel("XRS15 Ratio $(0.5-4\;\AA)$/$(1-8\;\AA)$")
ax.set_xlabel("Temperature [K]")
ax.set_xscale("log")
ax.set_yscale("log")
plt.legend()
plt.show()

fig,ax = plt.subplots(figsize=(7, 6))
plt.plot(1e6*resptmk15,resprat15/resprat13,marker=None,lw=2)
ax.set_ylabel("XRS15 Ratio / XRS13 Ratio")
ax.set_xlabel("Temperature [K]")
ax.set_xscale("log")
plt.show()
```
